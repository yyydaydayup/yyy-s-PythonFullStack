```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y = list(np.random.randint(0, 10, size=(10,)))
plt.plot(y)  # 默认横坐标为range(0,len(y))
```

```python
plt.plot(range(10,20) ,y)
```

## 笔记
1. plt.plot()可以只传Y轴的值，那么X轴默认为为range(0,len(Y))
2. plt.plot()的x和y不能作为关键字参数来传递，之鞥呢作为位置参数来传递。

```python
my_data = {
    'a':range(10, 20),
    'b':[np.random.randint(0,10) for x in range(10)]
}
plt.plot('a','b',"",data=my_data)
```

```python
my_df=pd.DataFrame(data=my_data)
plt.plot('a','b','m.:',data=my_df)
```
