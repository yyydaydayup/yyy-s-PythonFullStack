```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import font_manager
font = font_manager.FontProperties(fname=r"C:\\Windows\\Fonts\\msyh.ttc",size=15)
```

```python
scores = pd.read_csv('scores.csv').drop(columns=['num','class'])
avg = scores.mean()

values = list(avg.values)
values.append(values[0])
subject = list(avg.index)
theta = np.linspace(0, np.pi*2, len(avg)+1)

plt.polar(theta, values)
plt.xticks(theta, subject)
plt.fill(theta, values)
```
