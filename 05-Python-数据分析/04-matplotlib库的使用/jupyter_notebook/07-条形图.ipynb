## 01.普通条形图

```python
movies = {
    "流浪地球":40.78,
    "飞驰人生":15.77,
    "疯狂的外星人":20.83,
    "新喜剧之王":6.10,
    "廉政风云":1.10,
    "神探蒲松龄":1.49,
    "小猪佩奇过大年":1.22,
    "熊出没·原始时代":6.71
}
x = list(movies.keys())
y = list(movies.values())
plt.figure(figsize=(15,5))
# 方式一；直接传入列表
# plt.bar(x,y,width=0.5,color='red',edgecolor='green')
# 方式二：传入DataFrame对象或者字典
movies_df = pd.DataFrame(data={"names":list(movies.keys()),"tickets":list(movies.values())})
plt.bar("names", "tickets",data=movies_df)

plt.xticks(fontproperties=font,size=12)
plt.yticks(range(0, 45, 5),['%d亿'%tickets for tickets in range(0, 45, 5)],fontproperties=font,size=12)
plt.show()
```

## 03.分组条形图

```python
movies = {
    "流浪地球":[2.01,4.59,7.99,11.83,16],
    "飞驰人生":[3.19,5.08,6.73,8.10,9.35],
    "疯狂的外星人":[4.07,6.92,9.30,11.29,13.03],
    "新喜剧之王":[2.72,3.79,4.45,4.83,5.11],
    "廉政风云":[0.56,0.74,0.83,0.88,0.92],
    "神探蒲松龄":[0.66,0.95,1.10,1.17,1.23],
    "小猪佩奇过大年":[0.58,0.81,0.94,1.01,1.07],
    "熊出没·原始时代":[1.13,1.96,2.73,3.42,4.05]
}

movies_df = pd.DataFrame(data=movies)
movies_df
```

```python
plt.figure(figsize=(15,5))
xticks=np.arange(0,len(movies_df.columns))
width = 0.15

plt.xticks(xticks,movies_df.columns,fontproperties=font,size=12)

day_1 = movies_df.iloc[0]
plt.bar(xticks-2*width,day_1,width=width,label='第1天票房')
for index,ticket in enumerate(day_1):
    plt.annotate('%.1f'%ticket, xy=(xticks[index]-2*width,ticket),xytext=(xticks[index]-2*width-0.1, ticket+0.1))

day_2 = movies_df.iloc[1]
plt.bar(xticks-1*width,day_2,width=width,label='第2天票房')
for index,ticket in enumerate(day_2):
    plt.annotate('%.1f'%ticket, xy=(xticks[index]-1*width,ticket),xytext=(xticks[index]-1*width-0.1, ticket+0.1))

day_3 = movies_df.iloc[2]
plt.bar(xticks-0*width,day_3,width=width,label='第3天票房')
for index,ticket in enumerate(day_3):
    plt.annotate('%.1f'%ticket, xy=(xticks[index]-0*width,ticket),xytext=(xticks[index]-0*width-0.1, ticket+0.1))

day_4 = movies_df.iloc[3]
plt.bar(xticks+1*width,day_4,width=width,label='第4天票房')
for index,ticket in enumerate(day_4):
    plt.annotate('%.1f'%ticket, xy=(xticks[index]+1*width,ticket),xytext=(xticks[index]+1*width-0.1, ticket+0.1))

day_5 = movies_df.iloc[4]
plt.bar(xticks+2*width,day_5,width=width,label='第5天票房')
for index,ticket in enumerate(day_5):
    plt.annotate('%.1f'%ticket, xy=(xticks[index]+2*width,ticket),xytext=(xticks[index]+2*width-0.1, ticket+0.1))

    


plt.title("春节前5天电影票房记录",fontproperties=font)
plt.xlabel('电影',fontproperties=font)
plt.ylabel('票房（亿）',fontproperties=font)
font.set_size(12)
plt.legend(prop=font)



plt.grid()
plt.show()
```
