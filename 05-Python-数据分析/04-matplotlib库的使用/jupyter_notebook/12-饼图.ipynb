```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import font_manager
font = font_manager.FontProperties(fname=r"C:\\Windows\\Fonts\\msyh.ttc",size=15)
```

```python
oses = {
    'windows7':60.86,
    'windows10': 18.46,
    'windows8': 3.61,
    'windows xp': 10.3,
    'mac os': 6.78,
    '其他': 1.12
}

_,texts,autotexts = plt.pie(oses.values(),labels=oses.keys(),autopct='%.2f%%',textprops={'fontproperties':font,},explode=(0,0.1,0,0,0,0))
for autotext in autotexts:
    autotext.set_color('w')
    autotext.set_size(12)
plt.show()
```
